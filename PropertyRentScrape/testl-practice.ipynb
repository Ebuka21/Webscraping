### Webscrape function

```python
location = input('Which location are you searching for: ')
max_prize = input('State your max price: ')
bed_number = input('how many bedrooms: ')
page_num = 1

query = f'https://nigeriapropertycentre.com/for-rent/lagos/{location}?bedrooms={bed_number}&maxprice={max_prize}&selectedLoc=1&q=for-rent+lagos+{location}+{bed_number}+bedrooms+maxprice+{max_prize}&page={page_num}'
```

```python
query
```

```python
house_request = requests.get(query)
house_soup = bs4.BeautifulSoup(house_request.text, 'lxml')
```

```python
x = len(house_soup.select(".wp-block-body"))
#x
```

```python
f = 'FULL HOUSE'
print(f.lower())
```

```python
#for loop to go through all the different houses on a page
#use the strip command to remove space and letters
#use the replace command to remove comma from price and covert it to int

for house in house_soup.select(".wp-block-body"):
    house_dict['location'].append(house.address.text.strip(' \xa0'))
    house_dict['price'].append(int(house.select(".price")[1].text.replace(',','')))
    house_dict['link'].append(house.a['href'])
```

```python

info =  house_soup.select(".wp-block-body")
test_dict = {}
for n in range(len(info)):

    p = ['kitchen', 'shared', 'Prepaid Meter', 'sharing', 'shared', 'parlour', 'mini flat']
    
    test_dict['hd'+str(n)] = {
        'location':info[n].address.text.strip(' \xa0'), 
        'price':int(info[n].select(".price")[1].text.replace(',','')), 
        'link':'https://nigeriapropertycentre.com/'+info[n].a['href'],
        'features':[]
    }
    
    new_url = 'https://nigeriapropertycentre.com/' + info[n].a['href']
    mini_req = requests.get(new_url)
    mini_soup = bs4.BeautifulSoup(mini_req.text, 'lxml')
    a = []
    for p in p:
        if p in mini_soup.select(".tab-body")[0].text:
            test_dict['hd'+str(n)]['features'].append(p)
    
```
